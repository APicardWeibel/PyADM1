```python
import pandas as pd
d_script = pd.read_csv('src/dynamic_out.csv')
d_func = pd.read_csv('src/dynamic_out_func.csv')
d_mat = pd.read_csv('src/Matlabout_dyn.csv')

d_func = d_func.loc[0:(d_func.shape[0] - 2)]
d_script = d_script.loc[0:(d_script.shape[0] - 2)]
```

```python
import matplotlib.pyplot as plt
import numpy as np

for col_name in d_func.columns :
    print(f"Prediction {col_name}")

    script_val = d_script[col_name].to_numpy()
    func_val = d_func[col_name].to_numpy()
    mat_val = d_mat[col_name].to_numpy()
    #plt.plot(script_val, 'r')
    #plt.plot(func_val, 'b')
    #plt.plot(mat_val, 'y')

    print(f"Script to Func: {(np.mean( ((script_val - func_val) / mat_val)**2) ) ** (1/2) * 100} %")
    print( f"Script to Mat: {(np.mean( ((script_val - mat_val) / mat_val)**2) ) ** (1/2) * 100} %")
    print( f"Mat to Func: {(np.mean( ((mat_val - func_val) / mat_val)**2) ) ** (1/2) * 100} %")
    print('')
```
